```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/drive/MyDrive/Dataset_Final_Project/'

df_Cust = pd.read_excel(f'{path}Customers.xlsx')
df_Ord = pd.read_excel(f'{path}Orders.xlsx')
df_ProdCat = pd.read_excel(f'{path}ProductCategory.xlsx')
df_Prod = pd.read_excel(f'{path}Products.xlsx')

print('==Customers==')
display(df_Cust.head())
print('\n==Order==')
display(df_Ord.head())
print('\n==Product_Category==')
display(df_ProdCat.head())
print('\n==Product==')
display(df_Prod.head())
```

```python
print('==Customers==')
df_Cust.info()
print('\n==Order==')
df_Ord.info()
print('\n==Product_Category==')
df_ProdCat.info()
print('\n==Product==')
df_Prod.info()
```

```python
checks = {
    'Customers': (df_Cust, 'CustomerID'),
    'Orders': (df_Ord, 'OrderID'),
    'ProductCategory': (df_ProdCat, 'CategoryID'),
    'Products': (df_Prod, 'ProdNumber')
}

for table_name, (df, col) in checks.items():
    total_rows = len(df)
    unique_values = df[col].nunique()
    null_values = df[col].isna().sum()
    is_valid_pk = (total_rows == unique_values) and (null_values == 0)

    print(f'Tabel: {table_name}')
    print(f'  Kolom PK      : {col}')
    print(f'  Total baris   : {total_rows}')
    print(f'  Nilai unik    : {unique_values}')
    print(f'  Nilai kosong  : {null_values}')
    print(f'  Valid sebagai Primary Key? : {is_valid_pk}')
    print('-' * 40)
```

# **Relationship antar Dataset**

| Origin Table | Column (FK) | Objective Table | Column (PK) | Relationship Type
|-------------|-------------|-------------|-------------|-------------|
| Orders | CustomersID | Customers | CustomersID | One-to-Many (1-M) |
| Orders | ProdNumber | Products | Prodnumber | Many-to-one (M-1) |
| Products | Category | Product Category | CategoryID | Many-to-One (M-1) |

# **Creating New Master Dataset Based on Primary Dataset**

```python
# Join Orders with Customers

df_Ord_Cust = df_Ord.merge(df_Cust, on='CustomerID', how='inner')

df_Ord_Cust.info()
print('\n')
display(df_Ord_Cust.head())
```

```python
# Join df_Ord_Cust with Products

df_Ord_Cust_Prod = df_Ord_Cust.merge(df_Prod, on='ProdNumber', how='inner')

df_Ord_Cust_Prod.info()
print('\n')
display(df_Ord_Cust_Prod.head())
```

```python
# Join df_Ord_Cust_Prod with Product Category

df_Master = df_Ord_Cust_Prod.merge(df_ProdCat, left_on='Category', right_on='CategoryID', how='inner')

df_Master.info()
print('\n')
display(df_Master.head())
```

```python
# SUM Sales_total
df_Master['Sales_total'] = df_Master['Quantity'] * df_Master['Price']

# Rename 8 Columns
df_Final = df_Master[[
    'CustomerEmail',
    'CustomerCity',
    'Date',
    'Quantity',
    'ProdName',
    'Price',
    'CategoryName',
    'Sales_total'
]].copy()

df_Final.columns = [
    'cust_email',
    'cust_city',
    'order_date',
    'order_qty',
    'product_name',
    'product_price',
    'category_name',
    'total_sales'
]

# ASCED by Order_date
df_Final = df_Final.sort_values(by=['order_date', 'order_qty'], ascending=True).reset_index(drop=True)

print('Master Table Final')
df_Final.info()
print('\n')
display(df_Final.head(10))
print(f'\nTotal rows: {len(df_Final)}')
```

```python
df_Final.to_csv('Master_Sales_Data.csv', index=False)
```
